````python
scale = 0.75

for ann in anns:
    ann['area'] = ann['area']*scale**2
    ann['bbox'] = [i*scale for i in ann['bbox']]
    ann['keypoints'] = np.reshape(ann['keypoints'],(-1,3))*np.array([scale, scale, 1])
    ann['segmentation'] = ann['segmentation']*scale
    
````

# Concatenate labels as tensors

````python
bbox_list = tf.concat([tf.constant([ann['bbox']]) for ann in anns], axis=1)
keypoints_list = tf.concat([tf.constant(np.array([ann['keypoints']]).T) for ann in anns], axis=1)
seg_list = tf.concat([tf.constant([ann['segmentation'][0]]) for ann in anns], axis=1)
crowd_list = tf.concat([tf.constant([[ann['iscrowd']]]) for ann in anns] ,axis=1)
````

````python
sess = tf.Session()

bbox, kpts, segs, crowds = sess.run([bbox_list, keypoints_list, seg_list, crowd_list])
````
